```python
import pandas as pd
import numpy as np
import spacy
from spacy import displacy
from spacy.tokens import Token

# Load the language model
nlp = spacy.load("en_core_web_sm")
```

```python
def pos_parser(sentence):
    doc = nlp(sentence)
    count = 0
    sub_strings = ['I', 'i', 'You','you', 'Me', 'me', 'Mine','mine','your','Your', 'Myself', 'myself','It','it','That','that']

    pos_token = []
    for token in doc:
        if str(token.pos_) =='PRON' or str(token.pos_) =='PROPN':       
            pos_token.append(str(token.text))

        else:
            continue

    for token in doc:
        if str(token.pos_) =='VERB':
            if set(sub_strings).intersection(set([child.text for child in token.children])):
                continue            
            else:
                if set([child.text for child in token.children]).intersection(set(pos_token)):
                    #print(str(token.pos_), str(token.tag_), str(token.text), str(token.dep_), str(token.head.pos_), str([child.text for child in token.children]))
                    count = count + 1


                else:
                    continue        

        else:
            continue
    
    return count
```

```python
df = pd.read_csv('FN_review.csv')
```

```python
review_df = df['review']
```

```python
# data copy
fn_score = df.copy()

# Pos-tagging, Dependency Parsing
sentences = range(review_df.shape[0])

appended_data = []
for idx, value in enumerate(sentences):
    review_list = review_df.iloc[idx]
    data = pos_parser(review_list)
    appended_data.append(data)

fn_score.insert(7,'spoil_score', appended_data)
```
