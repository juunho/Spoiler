```python
import re
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer, PorterStemmer
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
from tqdm import tqdm
```

### Data Load

```python
df = pd.read_csv('IMDB/dataset/spacy-pre-sum_syn.csv')
```

```python
df
```

```python
df['review'].replace('', np.nan, inplace=True)
df.dropna(subset=['review'], inplace=True)

df['summary'].replace('', np.nan, inplace=True)
df.dropna(subset=['summary'], inplace=True)


df['synopsis'].replace('', np.nan, inplace=True)
df.dropna(subset=['synopsis'], inplace=True)
```

```python
df_1 = df.iloc[1:150000]
df_2 = df.iloc[150000:]
```

```python
df_1
```

```python
similarityValue = []

for i in tqdm(range(df_1.count()[0])):
    sentence_1 = nlp(df_1.iloc[i][3])
    sentence_2 = nlp(df_1.iloc[i][4])
    similarityValue.append(sentence_1.similarity(sentence_2))

df_1['summary_sim'] = similarityValue
```

```python
similarityValue = []

for i in tqdm(range(df_2.count()[0])):
    sentence_1 = nlp(df_2.iloc[i][3])
    sentence_2 = nlp(df_2.iloc[i][4])
    similarityValue.append(sentence_1.similarity(sentence_2))

df_2['summary_sim'] = similarityValue
```

```python
# review - synopsis similarity (1)
similarityValue = []

for i in tqdm(range(df_1.count()[0])):
    sentence_1 = nlp(df_1.iloc[i][3])
    sentence_2 = nlp(df_1.iloc[i][5])
    similarityValue.append(sentence_1.similarity(sentence_2))

df_1['synopsis_sim'] = similarityValue
```

```python
# review - synopsis similarity (2)
similarityValue = []

for i in tqdm(range(df_2.count()[0])):
    sentence_1 = nlp(df_2.iloc[i][3])
    sentence_2 = nlp(df_2.iloc[i][4])
    similarityValue.append(sentence_1.similarity(sentence_2))

df_2['synopsis_sim'] = similarityValue
```

```python
df_1.to_csv('IMDB/dataset/sim1-spacy-pre-unq.csv', index= False)
```

```python
df_2.to_csv('IMDB/dataset/sim2-spacy-pre-unq.csv', index= False)
```
